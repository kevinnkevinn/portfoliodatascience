## Extracting Feature columns and Target Columns

```python
X = data.iloc[:, 0:8]
y = data[q9]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, stratify = y, random_state= 0)
```

```python
print("Numbers of Train Dataset:", len(X_train))
print("Numbers of Test Dataset:", len(X_test))
```

## Machine Leanring

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_lr = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

labels = ['Beginner','Intermediate']
sns.heatmap(confusion_matrix(y_test, y_pred_lr), xticklabels = labels, yticklabels = labels, annot = True, fmt='d', cmap = "BuPu")
print(classification_report(y_test, y_pred_lr))
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred_nb)}\n')
sns.heatmap(confusion_matrix(y_test, y_pred_nb), xticklabels = labels, yticklabels = labels, annot = True, fmt='d', cmap = "YlGnBu")
print(f'Classification Report: \n {classification_report(y_test, y_pred_nb)}')
```

## Support Vector Classifier (SVC)

**Support Vector Classifier** is a supervised machine learning technique that is commonly employed for classification applications. The SVC algorithm operates by mapping data points to a high-dimensional space and then locating the best hyperplane that separates the data into two groups.

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
y_pred_svc = svc.predict(X_test)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred_svc)}\n')
sns.heatmap(confusion_matrix(y_test, y_pred_svc), xticklabels = labels, yticklabels = labels, annot = True, fmt='d', cmap = "Greens")
print(f'Classification Report: \n {classification_report(y_test, y_pred_svc)}')
```

## Decision Tree

**Decision Tree** is a tree structure that looks like a flowchart. Each internal node is a feature or attribute, each branch is a decision rule, and each leaf node is the result. The node at the very top of a decision tree is called the "root node." It learns to divide based on the value of an attribute. It divides the tree into sections in a way called "recursively." This structure, which looks like a flowchart, helps you decide what to do. It looks like a flowchart diagram, which is a good way to show how people think. Because of this, decision trees are easy to read and understand.

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred_dt)}\n')
sns.heatmap(confusion_matrix(y_test, y_pred_dt), xticklabels = labels, yticklabels = labels, annot = True, fmt='d', cmap ="OrRd")
print(f'Classification Report: \n {classification_report(y_test,y_pred_dt)}')
```

## Random Forest

**Random Forest** is an algorithm for supervised machine learning that is based on learning in groups. Ensemble learning is a way to learn in which different types of algorithms or the same algorithm is used more than once to make a stronger prediction model. The random forest algorithm combines several algorithms of the same type, called **decision trees.** This makes a forest of trees, which is why the algorithm is called **random forest.** When it comes to data analysis, the random forest approach is versatile enough to be utilized for both regression and classification.

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred_rf)}\n')
sns.heatmap(confusion_matrix(y_test, y_pred_rf), xticklabels = labels, yticklabels = labels, annot = True, fmt='d', cmap = "GnBu")
print(f'Classification Report: \n {classification_report(y_test, y_pred_rf)}')
```

## Deep Learning

The definition of deep learning can be interpreted as a technique in machine learning that enables a computer system or machine to function like a natural human, specifically by studying situations with specific learning or programming.

Additionally, the development of technology that is dependent on artificial intelligence or AI is made significantly easier through the use of deep learning (AI). During the process of deep learning, a computer will analyze a variety of models and categorize their respective duties depending on the data that it has acquired. It's possible for the data to be in the form of photos, text, or even sound. In point of fact, when dealing with enormous amounts of data, an even higher degree of precision is achieved.

```python
from tensorflow.keras.utils import to_categorical
from keras.layers import Input,Dense
from keras import Model
```

```python
y = data['Knowledge Level'].map({'Beginner':0,
                               'Intermediate':1})
y = to_categorical(y)
```

The map() function returns a map object (an iterator) containing the results of applying the given function to each item of an iterable.

```python
input = Input(shape=8)
x = Dense(32, activation='relu')(input)
output = Dense(2, activation='softmax')(x)

model = Model(inputs = input, outputs = output)
model.summary()
```

```python
model.compile(optimizer ='adam', loss='binary_crossentropy', metrics = ['acc'])
history = model.fit(X, y, validation_split = 0.1,epochs = 45)
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Accuracy', fontsize = 18)
plt.ylabel('Accuracy', fontsize = 14)
plt.xlabel('Epoch', fontsize = 14)
plt.legend(['Train', 'Validation'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss', fontsize = 18)
plt.ylabel('Loss', fontsize = 14)
plt.xlabel('Epoch', fontsize = 14)
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
```
